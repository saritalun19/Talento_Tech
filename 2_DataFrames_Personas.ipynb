# Archivo personas 1

```python
import pandas as pd

file_1="/content/drive/MyDrive/Sarita Luna/Componente tecnico/Excel/Taller 11 de Junio/Set Datos-Junio 11.xlsx - Data_1.csv"

df_persona_1=pd.read_csv(file_1, delimiter=",", header=1)

df_persona_1
```

```python
df_persona_1.head()
```

```python
df_persona_1.rename(columns={'Weight':'Peso', 'Height':'Altura', 'Age':'Edad', 'Gender':'Genero', 'Income':'Salario'}, inplace=True)
df_persona_1
```

```python
# Transformar los valores de peso y estatura
#Factores de conversion

lbs_to_kg=0.453592
estatura_to_mts=0.3048

#Convertir peso a KG
df_persona_1['Peso kg']=df_persona_1['Peso']*lbs_to_kg

#Convertir estatura a mts
df_persona_1['Altura mts']=df_persona_1['Altura']*estatura_to_mts

df_persona_1
```

```python
df_seleccionado_1=df_persona_1[['id', 'Peso kg', 'Altura mts','Genero','Edad']]
df_seleccionado_1
```
